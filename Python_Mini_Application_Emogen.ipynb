```python
History=[]
```

```python
def display_banner():
  """ Display a welcome banner """
  print("="*70)
  print("ðŸ™ Welcome to Emoji Mood Translator ðŸ™")
  print("="*70)
```

```python
def get_user_input() -> str:
  """Prompt the user for a mood and return it in lowercase."""
  mood=input("\nEnter how you feel today (happy, sad, joyful, angry, excited, tired, sleepy, confused, bored, scared, love, motivated, surprised, relaxed): "). strip().lower()
  while not mood:
    print("The space cannot be left idle")
    mood=input("Enter the mood again:").strip().lower()
  return mood
```

```python
def translate_mood(mood: str) -> str:
  """Translate the user's mood to an emoji. Return ðŸ¤” if nothing found"""
  global Mood_Dict
  return Mood_Dict.get(mood, "ðŸ¤”")
```

```python
from os import times
def log_mood(mood: str, emoji: str):
  """Mark the mood with timestamp"""
  global History
  timestamp=datetime.now().strftime("%Y-%m-%d %H:%M:%S")
  History.append({"time":timestamp, "mood": mood, "emoji": emoji})

  with open("mood_history.txt", "a") as f:
    f.write(f"{timestamp} - {mood} -> {emoji}\n")
```

```python
def show_history():
  """Display all past translations"""
  if not History:
    print("No mood history found")
  else:
    print("Mood History:")
    for entry in History:
      print(f"{entry['time']} - {entry['mood']} -> {entry['emoji']}")
```

```python
def get_message_for_mood(mood: str) -> str:
  """Return a message as per the user's mood"""
  if mood in Messages:
    return random.choice(Messages[mood])
  else:
    return "No message found"
```

```python
def suggest_similar_moods(user_mood: str):
  """Suggest moods similar to the one entered, depending on the initial letter"""
  global Mood_Dict
  similar = [m for m in Mood_Dict if m.startswith(user_mood[0]) and m != user_mood]
  if similar:
    print("Did you mean any of the moods mentioned here?")
    for s in similar:
      print(".", s)
```

```python
def main():
  """Main program loop"""
  display_banner()
  while True:
    print("\nMenu:")
    print("1) Translate a mood")
    print("2) Show mood history")
    print("3) Project Information")
    print("4) Exit")

    choice=input("Enter your choice: ")

    if choice=="1":
      user_mood=get_user_input()
      emoji=translate_mood(user_mood)
      if emoji == "ðŸ¤”":
        print("\nSorry, I don't know that mood")
        suggest_similar_moods(user_mood)
      else:
        message = get_message_for_mood(user_mood)
        print("\n----------------------------")
        print(f"\n{user_mood.capitalize()} is {emoji}")
        print(f" Message: {message}")
        print("----------------------------\n")
        log_mood(user_mood, emoji)

    elif choice =="2":
      show_history()

    elif choice =="3":
      print("\n Project Information:")
      print("This is a mini python app")
      print("It demonstrates the use of dictionaries, loops, functions,")
      print("input validation and simple file handling in python")

    elif choice =="4":
      print("\nThank you")
      print("Have a great day ahead!ðŸ˜Š")
      break

    else:
      print("\nInvalid choice. Please select 1,2,3 or 4.")

if __name__=="__main__":
  main()
```
